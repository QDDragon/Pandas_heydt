# Seaborn plotting system

## strip plot

```python
"""
A strip plot is a scatter plot where one of the variables is categorical.
Strip plots can be combined with other plots to provide additional information. 
For example, a box plot with an overlaid strip plot is similar to a violin plot 
because some additional information about how the underlying data is distributed becomes visible

Seaborns swarm plot
is virtually identical to a strip plot except that it prevents data points from
overlapping
"""
import seaborn as sb
sb.stripplot(x=dataset['salary'])
```

```python
sb.stripplot(x=dataset['sex'], y = dataset['salary'], data = dataset)
# visualizes the salary variable per gender
```

```python
sb.stripplot(x=dataset['sex'], y = dataset['salary'], data = dataset, jitter=0.001)
# jitter: su bien dong, chap chon
#draw in vertical
```

```python
sb.stripplot(y=dataset['sex'], x = dataset['salary'], data = dataset, jitter=0.001)
# draw in horizontal
```

```python
sb.stripplot(x=dataset['discipline'], y = dataset['salary'], data = dataset, jitter =1)
"""
Discipline B has a bigger range
and higher packages compared to discipline A
"""
```

```python
dataset[['discipline', 'salary']]
```

```python
sb.stripplot(x=dataset['rank'], y = dataset['salary'], data = dataset, jitter = True)
```

```python
sb.stripplot(x='sex', y='salary', hue='rank', data=dataset, jitter=True)
"""
The hue attribute is used to determine the legend attribute.
"""
```

## Box plot

```python
"""
combine a box plot and strip plot to give more information on the
generated plot.

As shown, the Male category has a higher median salary, maximum salary, 
and range compared to the Female category.
"""
import numpy as np
sb.boxplot(x='salary', y = 'sex', data=dataset, whis=np.inf)
"""
By default,the whiskers extend to 1.5 * the interquartile range
(check Theories whisker in oneNote)
Make them extend to 2.0 * IQR: whis=2.0
Show the 5th and 95th percentiles: whis=[5, 95]
Show min and max values: whis=[0, 100]
"""

sb.stripplot(x='salary', y = 'sex', data=dataset,
            jitter= True, color='0.02')
# Draw data on top of boxplot
```

```python
sb.boxplot(x=dataset.salary)
```

```python
#box plot per rank
sb.boxplot(x='rank', y = 'salary', data= dataset)
```

```python
#box plot per rank using stype (set3)
sb.boxplot(x='rank', y ='salary', hue='sex', data=dataset,
          palette='Set3')
```

```python
# box plot per rank + swarm 
sb.boxplot(x = 'rank', y = 'salary', data=dataset)
sb.swarmplot(x = 'rank', y = 'salary', data=dataset,
color='0.25')
```

```python
## Swarm plot
"""
A swarm plot is used to visualize different categories; it gives a clear
picture of a variable distribution against other variables. 

For instance,
the salary distribution per gender and per profession indicates that the
male professors have the highest salary range. 

Most of the males are full professors, then associate, and then assistant professors. 
There are more male professors than female professors, but there are more female
associate professors than male associate professors
"""
sb.swarmplot(x='sex', y='salary', hue = 'rank', data=dataset, 
            palette='Set2', dodge=False)
```

```python
sb.swarmplot(x='sex', y='salary', hue = 'rank', data=dataset, 
            palette='Set2', dodge=True)
```

## Joint Plot

```python
"""
A joint plot combines more than one plot to visualize the selected patterns
"""
sb.jointplot(x='salary', y='service', data=dataset)
```

```python
sb.jointplot('salary', 'service', data=dataset, kind='reg')
```

```python
sb.jointplot('salary', 'service', data=dataset, kind='hex')
```

```python
sb.jointplot('salary', 'service', data=dataset, kind='kde')
```

```python
from scipy.stats import spearmanr 
sb.jointplot('salary', 'service', data=dataset, stat_func=
spearmanr )
"""
Use the Spearman Rank Correlation Coefficient (R) to measure 
the relationship between two variables 
where one or both is not normally distributed
"""
# Correlation (Pearson, Kendall, Spearman)
# https://www.youtube.com/watch?v=DE58QuNKA-c
```

```python
sb.jointplot('salary', 'service',
data=dataset).plot_joint(sb.kdeplot, n_levels=6)
```

```python
sb.jointplot('salary', 'service',
data=dataset).plot_joint( sb.kdeplot,n_levels=6).plot_marginals(sb.rugplot)
```
